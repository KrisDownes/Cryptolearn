```python
def hex2dec(x):
    import numpy as np
    x = str(x).lower()
    d = {'a':10,'b':11,'c':12,'d':13,'e':14,'f':15}
    L = len(x)
    coeff, expon = [],[]
    t = []
    for i , n in enumerate(x):
        coeff.append(n)
        expon.append(int(L-i-1))
    for i in range(len(coeff)):
        if coeff[i] in d.keys():
            coeff[i] = d[coeff[i]]
            coeff[i] = int(coeff[i])
        else :
            coeff[i] = int(coeff[i])
    for i in range(len(coeff)):
        t.append(coeff[i]*(16**expon[i]))
    return np.sum(t)
```

```python
def decToBin(dec):
    cnt = 0
    binary = 0
    while dec > 0:
        binary = int((dec %2))*int((10**cnt)) + binary
        dec = (dec//2)
        cnt +=1
    return binary
```
